12) Using the dataframes created in question 12, perform left join to combine values in the columns 'Brand' and 'ID

```python
import pandas as pd

# First DataFrame (left table)
df1 = pd.DataFrame({
    'ID': [101, 102, 103, 104, 105],
    'Brand': ['Apple', 'Canon', 'Samsung', 'Nikon', 'Sony'],
    'Product': ['iPhone', 'DSLR', 'SmartPhone', 'DSLR', 'SmartTV']
})

# Second DataFrame (right table)
df2 = pd.DataFrame({
    'ID': [101, 102, 104, 103, 105],
    'Brand': ['Apple', 'Canon', 'Samsung', 'Nikon', 'Sony'],
    'Quantity_sold': [234, 344, 345, 262, 356]
})

# Perform left join on 'ID' and 'Brand'
merged_df = pd.merge(df1, df2, how='left', on=['ID', 'Brand'])
print(merged_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Step 2: Create a strip plot for sepal length across different species
plt.figure(figsize=(8, 6))
sns.stripplot(x='species', y='sepal_length', data=iris, jitter=True, palette='Set2')

# Step 3: Add title and labels
plt.title('Sepal Length Distribution Across Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')

# Step 4: Show the plot
plt.tight_layout()
plt.show()
```

20) Plot a heat map for the correlation matrix using dataset iris

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Load the Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Step 2: Compute the correlation matrix for the numerical columns
corr_matrix = iris.drop(columns=['species']).corr()

# Step 3: Plot the heatmap using seaborn
plt.figure(figsize=(8, 6))  # Set the figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Step 4: Add a title and display the plot
plt.title('Correlation Matrix Heatmap - Iris Dataset')
plt.tight_layout()
plt.show()
```

21) Using the 'diamonds' data available in the library seaborn, consider the price for the type of cut. Display the data using the following plots:
                        <ol>
                            <li> Multiple boxplot</li>
                            <li> Violin Plot </li>
                            <li> Boxen Plot </li>
                          

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Step 2: Set up the figure for plotting multiple plots
plt.figure(figsize=(18, 6))

# 1. Multiple Boxplot
plt.subplot(1, 3, 1)
sns.boxplot(x='cut', y='price', data=diamonds, palette='Set2')
plt.title('Boxplot of Price by Cut')

# 2. Violin Plot
plt.subplot(1, 3, 2)
sns.violinplot(x='cut', y='price', data=diamonds, palette='Set2')
plt.title('Violin Plot of Price by Cut')

# 3. Boxen Plot
plt.subplot(1, 3, 3)
sns.boxenplot(x='cut', y='price', data=diamonds, palette='Set2')
plt.title('Boxen Plot of Price by Cut')

# Step 3: Adjust layout and display the plots
plt.tight_layout()
plt.show()

```
